### EDA and Feature Engineering - Flight Price Prediction

```python
# Importing Important Modules
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
# Creating DataFrame
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
# get the basic info about the dataset
df.info()
```

```python
df.head()
```

```python
# Splitting Date_of_journey column in to date, month, year
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
# Checking For Splitted Column
df.head()
```

```python
#Have some Extra Unwanted same column named as date
df.info()
```

```python
# Dropping unwanted Date Column
df.drop('date',axis=1,inplace=True)
```

```python
#checking for removed column
df.head()
```

```python
# Removing Date_of_journey Column because we splitted that column
df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
# Checking for successful removal of date_of_journey column
df.head()
```

```python
# checking for data types
df.info()
```

```python
# converting datatypes for newely created column
df['Date'] = df['Date'].astype('int64')
df['Month'] = df['Month'].astype('int64')
df['Year'] = df['Year'].astype('int64')
df.info()
```

```python
# Going further for another column
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split( )[0])
```

```python
# Splitting Hour and minutes seperately from the column and add seperate column for hour and min
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
df.head(2)
```

```python
# converting datatypes for the newely created column hour and min
df['Arrival_hour'] = df['Arrival_hour'].astype('int64')
df['Arrival_min'] = df['Arrival_min'].astype('int64')
df.info()
```

```python
# Drop Original Arrival_Time Column
df.drop('Arrival_Time',axis = 1, inplace = True)
df.info()
```

```python
df.head()
```

```python
# Going Forward for other column Dep_time
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
df.head(2)
```

```python
# Converting datatypes for dep_hour and dep_min
df['Dep_hour'] = df['Dep_hour'].astype('int64')
df['Dep_min'] = df['Dep_min'].astype('int64')
df.info()
```

```python
# Removing Original Dep_time Column
df.drop('Dep_Time',axis = 1, inplace = True)
df.head(2)
```

```python
df.info()
```

```python
df.head(2)
```

```python
# Dropping Route Column which is not needed because we have arriving and destination column
df.drop('Route', axis = 1, inplace = True)
df.head(2)
```

```python
# Going to forward to othe columns Total_Stops
df['Total_Stops'].unique()
```

```python
# Finding How many nan values in the column

df[df['Total_Stops'].isnull()]
```

```python
# Taking central endency to know which is best to replace that nan value
df['Total_Stops'].mode()
```

```python
# using map function we can replace nan value
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':4,'1 stop':3,'2 stops':2,'3 stops':1,'4 stops':0,np.nan:1})
df.head()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.head(2)
```

```python
# Spliting the columns and creating the new columns 
df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
# Created new columns for the duration hours and duration minutes
df['D-hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['D-minutes'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.head()
```

```python
df['D-hour'] = df['D-hour'].str.replace(r'[^0-9]', '', regex=True)
df['D-minutes'] = df['D-minutes'].str.replace(r'[^0-9]', '', regex=True)
```

```python
df['D-hour'] = df['D-hour'].fillna('0').astype('int64')
df['D-minutes'] = df['D-minutes'].fillna('0').astype('int64')
```

```python
df.info()
```

```python
# removing duration columns
df.drop('Duration', axis = 1, inplace = True)
df.info()
```

```python
# Converting All the Categories Features in to numerical features for ML by usin OneHotEncoding
# Ml works better numerical features 
# for onehotencoder we need to import sklearn module
from sklearn.preprocessing import OneHotEncoder
```

```python
# Defining Encoder in to a variable
encoder = OneHotEncoder()
```

```python
# Using fit_transform method converting category in to numerical value and make it as array using toarray method
encoded = encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
# takeing those array in to variable
df_encoded = pd.DataFrame(encoded,columns = encoder.get_feature_names_out())
```

```python
df_encoded
```

```python
# Adding those newely created categories in to new variable
df_v01 = pd.concat([df,df_encoded],axis = 1)
```

```python
df_v01.info()
```

```python
# Checking for null values
df_v01.isnull().sum()
```

```python
# Need to remove the categories columns
df_v01.drop(['Airline','Source','Destination'],axis = 1, inplace = True)
df_v01
```
